Importing the dependencies


```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection & Processing

```python
#loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
#loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns= breast_cancer_dataset.feature_names)
```

```python
#print tha first 5 rows of the dataframe
data_frame.head()
```

```python
#adding the target column to the data frame
data_frame['label']= breast_cancer_dataset.target
```

```python
#print last 5 rowa of the data frame
data_frame.tail()
```

```python
#numbers of rows and columns in the dataset
data_frame.shape
```

```python
#getting some information from the data
data_frame.info()
```

```python
#checking for missing values
data_frame.isnull().sum()
```

```python
#statistical measures about the data
data_frame.describe()
```

```python
#checking the distribution the Target Variable
data_frame['label'].value_counts()
```

1--> Benign
0--> Malignant

```python
data_frame.groupby('label').mean()
```

Separating the features and atrget

```python
x = data_frame.drop(columns='label', axis=1)
y = data_frame['label']
```

```python
print(x)
```

```python
print(y)
```

Spliting the data into training data & Testing data

```python
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=2)
```

```python
print(x.shape,x_train.shape, x_test.shape)
```

Model Trainning
Logistic Regression

```python
model = LogisticRegression()
```

```python
#training the Logistic Regression model using training data
model.fit(x_train, y_train)
```

Model Evaluation
Accuracy score

```python
#accuracy in training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(y_train, x_train_prediction)
```

```python
print('Accuracy on training data=', training_data_accuracy)
```

```python
#accuracy in test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(y_test, x_test_prediction)
```

```python
print('Accuracy on test data=', test_data_accuracy)
```

Building a Predictive Data

```python
input_data =(19.69,21.25,130,1203,0.1096,0.1599,0.1974,0.1279,0.2069,0.05999,0.7456,0.7869,4.585,94.03,0.00615,0.04006,0.03832,0.02058,0.0225,0.004571,23.57,25.53,152.5,1709,0.1444,0.4245,0.4504,0.243,0.3613,0.08758)

# change the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting one datapoint
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
prediction = model.predict(input_data_reshaped)
print(prediction)

if(prediction[0] == 0):
  print('The Breast cancer is Malignant')
else:
  print('The Breast cancer is Benign')
```
